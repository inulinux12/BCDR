## Parameter exploration using regularization

One of the regularization parameters that Caffe allows us to modify is the weight decay which is a tradeoff between the large weights and the function to minimize. The value used to train ImageNet was 0.0005, however we are going to find a suitable parameter between 5*10^-4 to 5*10^-1. The following are the accuracy results we got from the different training and their test results.

```python
import numpy as np
results = []
results.append(np.genfromtxt('accuracy3.csv').tolist())
results.append(np.genfromtxt('output-training3-1.txt-accuracy.csv').tolist())
results.append(np.genfromtxt('output-training3-2.txt-accuracy.csv').tolist())
results.append(np.genfromtxt('output-training3-3.txt-accuracy.csv').tolist())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15,10))
plt.plot([0.0005,0.005,0.05,0.5], results, '*')
plt.xscale('log')
plt.xlabel("Weight decay")
plt.ylabel("Accuracy")
```

```python
x1 = np.linspace(0, 10000, 21)

plt.figure(figsize=(15,10))
plt.subplot(4, 1, 1)
plt.plot(x1, results[0], 'yo-')
plt.title('5*10^-4')
plt.ylabel('Accuracy')

plt.subplot(4, 1, 2)
plt.plot(x1, results[1], 'ro-')
plt.title('5*10^-3')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')

plt.subplot(4, 1, 3)
plt.plot(x1, results[2], 'go-')
plt.title('5*10^-2')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.tight_layout()

plt.subplot(4, 1, 4)
plt.plot(x1, results[3], 'bo-')
plt.title('5*10^-1')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

We will use the two last models with weight decay.

Test:

- 0.005(iteration 6000): 0.435625->0.675
- 0.05(iteration 10000): 0.47125->0.435625
