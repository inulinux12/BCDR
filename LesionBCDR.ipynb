## Parameter exploration using regularization

One of the regularization parameters that Caffe allows us to modify is the weight decay which is a tradeoff between the large weights and the function to minimize. The value used to train ImageNet was 0.0005, however we are going to find a suitable parameter between 5*10^-4 to 5*10^-1. The following are the accuracy results we got from the different training and their test results.

```python
import numpy as np
results = []
results.append(np.genfromtxt('accuracy3.csv').tolist())
results.append(np.genfromtxt('output-training3-1.txt-accuracy.csv').tolist())
results.append(np.genfromtxt('output-training3-2.txt-accuracy.csv').tolist())
results.append(np.genfromtxt('output-training3-3.txt-accuracy.csv').tolist())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Importación de Librerias

```python
#Some pretty and interesting functions

#import caffe
from IPython.display import display, Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
%matplotlib inline

import base64
import skimage.io

class Image(skimage.io.Image):
    def __init__(self, *args, **kwargs):
        self._width = kwargs.pop('width', '100%')
        skimage.io.Image.__init__(self, *args, **kwargs)
    def _repr_html_(self):
        out = '<img width="%s"' % self._width
        out += ' src="data:image/png;base64,' + \
               base64.b64encode(np.compat.asbytes(self._repr_png_())) + '"/>'
        return out


class side_by_side():
    def __init__(self, *frames):
        self.frames = frames

    def _repr_html_(self):
        s = ""
        for f in self.frames:
            s += "<div style='float: left;'>" + f._repr_html_() + "</div>"

        return s
```
